# Session 8: Data Visualization with Seaborn  

## Overview  
In this session, we use Python's Seaborn library to visualize and analyze a dataset containing information about students, such as their study time, grades, courses, gender, and more. Each exercise focuses on creating a specific type of plot to derive insights from the data.  

---

## Exercises  

1. **Lineplot: Study Time by Student Name**  
   Create a lineplot to visualize how study time varies by student and identify the student with the highest study time.  

2. **Histogram: Grade Distribution**  
   Create a histogram (histplot) to visualize the frequency distribution of grades and determine which grade range has the highest frequency of students.  

3. **ECDF Plot: Cumulative Grade Distribution**  
   Plot an ECDF (ecdfplot) to show the cumulative percentage of grades and calculate the percentage of students scoring below 85.  

4. **Stripplot: Grade Distribution by Course**  
   Create a stripplot to visualize individual grade distributions for each course and identify the course with the most spread in grades.  

5. **Swarmplot: Gender and Study Time**  
   Plot a swarmplot to analyze the relationship between gender and study time and determine which gender has a higher average study time.  

6. **Pointplot: Average Grade by Course**  
   Create a pointplot to compare the average grade across courses and identify the course with the highest average grade.  

---


```python
#Importing libraries and loading data set

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Student Name': ['Amelie', 'Edgar', 'Jordi', 'Mar√ßal', 'Pep', 'Jaume', 'Marco', 
                     'Ludmila', 'Bastian', 'Marc', 'Gerardo', 'Javier', 'Frank', 'Julia'],
    'Course': ['Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science', 'Cloud Computing', 'Computer Vision', 
               'Artificial Intelligence', 'Python for Data Science'],
    'Grade': [85, 92, 78, 88, 91, 84, 73, 95, 90, 80, 85, 87, 89, 92],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Female'],
    'Study Time': [15, 20, 10, 8, 18, 14, 11, 16, 19, 13, 16, 12, 17, 15],
    'Age': [22, 21, 23, 22, 20, 24, 22, 21, 23, 21, 22, 22, 21, 20]
}

df = pd.DataFrame(data)
```

#### Exercise 1:

```python
# Plotting a line to show how much time each student spends studying.

# Set the figure size
plt.figure(figsize=(8, 4))

# Create a lineplot to visualize study time for each student
sns.lineplot(data=df, x='Student Name', y='Study Time', marker='o')

# Add title to chart
plt.title('Study Time by Student Name')

# Label x-axis
plt.xlabel('Student Name')

# Label y-axis
plt.ylabel('Study Time (Hours)')

# Adjust the layout to prevent overlapping elements
plt.tight_layout()

plt.show()
```

#### Exercise 2:

```python
# Creating a bar chart to visualize how many students fall into each grade range.

# Figure size
plt.figure(figsize=(8, 4))

# Create a histogram to show the distribution of grades with a density line (kde=True)
sns.histplot(data=df, x='Grade', bins=10, kde=True)

# Chart title
plt.title('Grade Distribution')

# X-axis label
plt.xlabel('Grade')

# y-axis label
plt.ylabel('Frequency')

# Adjust the layout to prevent overlapping elements
plt.tight_layout()

# Display the plot
plt.show()
```

#### Exercise 3:

```python
# Plotting cumulative percentages of grades and calculating the percentage of students with grades below 85.

# Figure size
plt.figure(figsize=(8, 4))

# Create an ECDF plot to show cumulative percentages for grades
sns.ecdfplot(data=df, x='Grade')

# Chart title
plt.title('ECDF of Grades')

# X-axis label
plt.xlabel('Grade')

# Y-axis label
plt.ylabel('ECDF')

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()

# Calculate the percentage of students scoring less than 85
percentage_less_than_85 = (df['Grade'] < 85).mean() * 100

# Print the result
print(f"Percentage of students scoring less than 85: {percentage_less_than_85:.2f}%")
```

#### Exercise 4:

```python
# Plotting individual grades for each course to see their spread.

# Figure size
plt.figure(figsize=(8, 4))

# Create a stripplot to show how grades are distributed for each course, with added jitter for clarity
sns.stripplot(data=df, x='Course', y='Grade', jitter=True)

# Chart title
plt.title('Grade Distribution by Course')

# X-axis label
plt.xlabel('Course')

# Y-axis label
plt.ylabel('Grade')

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

#### Exercise 5:

```python
# Showing the distribution of study times for males and females.

# Figure size
plt.figure(figsize=(8, 4))

# Create a swarmplot to show study time distribution for each gender
sns.swarmplot(data=df, x='Gender', y='Study Time')

# Chart title
plt.title('Study Time by Gender')

# X-axis label
plt.xlabel('Gender')

# Y-axis label
plt.ylabel('Study Time (Hours)')

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

#### Exercise 6:

```python
# Plotting the average grade for each course to compare performance.

# Figure size
plt.figure(figsize=(8, 4))

# Create a pointplot to show the average grade for each course
sns.pointplot(data=df, x='Course', y='Grade', estimator='mean')

# Chart title
plt.title('Average Grade by Course')

# X-axis label
plt.xlabel('Course')

# Y-axis label
plt.ylabel('Average Grade')

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```
